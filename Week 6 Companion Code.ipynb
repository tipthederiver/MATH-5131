### SIR Model

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

def SIRmodel(Y, t, b,g):
    x, y = Y
    dydx = [-b*x*y, b*x*y - g*y]
    return dydx

t = np.linspace(0,60,500)  # Max around 20
b,g = [.005,.2]

plt.figure(figsize=(12,8))

for ys in [20, 80, 200]:
    Y0 = [ys,  1]
    Y_sol = odeint(SIRmodel,Y0,t,args=(b,g))

    #Creates two subplots and unpacks the output array immediately
    X1 = Y_sol[:,0]
    X2 = Y_sol[:,1]

    xs = g/b

    plt.plot(X1,X2)
    plt.plot([xs,xs],[min(X2),max(X2)+20], )

plt.ylabel('Infected')
plt.xlabel('Susceptible')

#plt.legend(loc='center right')
```

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

def PPmodel(Y, t, a,b,c1,c2):
    x, y = Y
    dydx = [a*x - c1*x*y, c2*x*y - b*y]
    return dydx

t = np.linspace(0,1000,500)  # Max around 20

Y0 = [200,  80]

## Good parameters: 
#     [1,.5,.01,.005]     - Oscillating Popilation
#     [1,.5,.01,.001]     - Oscillating Popilation w/few preditors
#     [1,.5,.1,.001]      - Apparent Preditor Collapse
#     [.01,.5,.1,.001]    - Apparent Prey Collapse

a,b,c1,c2 = [.01,.5,.1,.001]

Y_sol = odeint(PPmodel,Y0,t,args=(a,b,c1,c2))

#Creates two subplots and unpacks the output array immediately
X1 = Y_sol[:,0]
X2 = Y_sol[:,1]

plt.figure(figsize=(12,8))

xs = b/c2
ys = a/c1

plt.plot(t,X1,label='x',color='c')
plt.plot(t,xs*np.ones(500),'--',color='c',label='x stable')
plt.plot(t,X2,label='y',color='r')
plt.plot(t,ys*np.ones(500),'--',color='r',label='y stable')
plt.xlabel('Time t')
plt.ylabel('Population')

plt.legend(loc='center right')
```

## Preditor Prey: DDT

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

def PPmodel(Y, t, a,b,c1,c2):
    x, y = Y
    dydx = [a*x - c1*x*y, c2*x*y - b*y]
    return dydx

t = np.linspace(0,20,500)  # Max around 20

Y0 = [200,  80]

## Good parameters: 
#     [1,.5,.01,.005]     - Oscillating Popilation
#     [1,.5,.01,.001]     - Oscillating Popilation w/few preditors
#     [1,.5,.1,.001]      - Apparent Preditor Collapse
#     [.01,.5,.1,.001]    - Apparent Prey Collapse

a,b,c1,c2 = [1,.5,.01,.005]

Y_sol = odeint(PPmodel,Y0,t,args=(a-.1,b+.1,c1,c2))
Y_solO = odeint(PPmodel,Y0,t,args=(a,b,c1,c2))

#Creates two subplots and unpacks the output array immediately
X1 = Y_sol[:,0]
X2 = Y_sol[:,1]

plt.figure(figsize=(12,8))

xs = (b+.1)/c2
ys = (a-.1)/c1


plt.plot(t,X1,label='x DDT',color='c')
plt.plot(t,xs*np.ones(500),'--',color='c',label='x stable')
plt.plot(t,X2,label='y DDT',color='r')
plt.plot(t,ys*np.ones(500),'--',color='r',label='y stable')


X1 = Y_solO[:,0]
X2 = Y_solO[:,1]
xs = b/c2
ys = a/c1


plt.plot(t,X1,label='x DDT',color='y')
plt.plot(t,xs*np.ones(500),'--',color='y',label='x stable')
plt.plot(t,X2,label='y DDT',color='y')
plt.plot(t,ys*np.ones(500),'--',color='y',label='y stable')

plt.xlabel('Time t')
plt.ylabel('Population')

plt.legend(loc='center right')
```

## Density Dependant Growth:

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

def PPDDmodel(Y, t, a,b,c1,c2,K):
    x, y = Y
    dydx = [a*x*(1-x/K) - c1*x*y, c2*x*y - b*y]
    return dydx

t = np.linspace(0,40,500)  # Max around 20

Y0 = [120,  80]

## Good parameters: 
#     [1,.5,.01,.005]     - Oscillating Popilation
#     [1,.5,.01,.001]     - Oscillating Popilation w/few preditors
#     [1,.5,.1,.001]      - Apparent Preditor Collapse
#     [.01,.5,.1,.001]    - Apparent Prey Collapse

a,b,c1,c2,K = [1,.5,.01,.005,1000]

Y_sol = odeint(PPDDmodel,Y0,t,args=(a,b,c1,c2,K))

#Creates two subplots and unpacks the output array immediately
X1 = Y_sol[:,0]
X2 = Y_sol[:,1]

plt.figure(figsize=(12,8))

xs = (b)/c2
ys = (a/c1)*(1-b/(K*c2)) 


plt.plot(t,X1,label='x DDT',color='c')
plt.plot(t,xs*np.ones(500),'--',color='c',label='x stable')
plt.plot(t,X2,label='y DDT',color='r')
plt.plot(t,ys*np.ones(500),'--',color='r',label='y stable')

plt.xlabel('Time t')
plt.ylabel('Population')

plt.legend(loc='upper right')
```
